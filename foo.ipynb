```python
import spacy
from spacy import displacy
```

```python
text = "My living room is 80ft by 35ft. How many couches can I fit in it?"
text = "How many 2ft by 7ft objects can fit in an 80ft by 35ft rectangle?"
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
displacy.render(doc, style='dep')
```
